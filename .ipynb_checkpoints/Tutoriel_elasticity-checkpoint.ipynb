### Fenics tutorial

```python
import pyvista
from dolfinx import mesh, fem, plot, io, default_scalar_type
from dolfinx.fem.petsc import LinearProblem
from mpi4py import MPI
import ufl
import numpy as np

L = 1.0
W = 0.2
mu = 1.0
rho = 1.0
delta = W/L
gamma = 0.4 * delta**2
beta = 1.25
lambda_ = beta
g = gamma

domain = mesh.create_box(
    MPI.COMM_WORLD,
    [np.array([0,0,0]), np.array([L,W,W])],
    [20, 6, 6],
    cell_type=mesh.CellType.hexahedron,   
)

# Definition de l espace des fonctions ou seront contenus les deplacements
V = fem.functionspace(domain, ("Lagrange", 1, (domain.geometry.dim,)))

# Boundary condition Dirichletbc
def clamped_boundary(x):
    return np.isclose(x[0],0)

# Selection des facettes
fdim = domain.topology.dim - 1 # les facettes sont en dimension 2 pour le bord
boundary_facets = mesh.locate_entities_boundary(domain, fdim, clamped_boundary) # retourne l ensemblde des facetts pour le bord encastre

u_D = np.array([0,0,0], dtype=default_scalar_type)
bc = fem.dirichletbc(u_D, fem.locate_dofs_topological(V, fdim, boundary_facets), V) # fem.locate.dofs_topological(V, fdim, boundary_facets) selectionne les degres de liberte des facettes

# Traction free verywhere else
T = fem.Constant(domain, default_scalar_type((0,0,0)))

ds = ufl.Measure("ds", domain=domain)

# Variational formulation
def epsilon(u):
    return ufl.sym(
        ufl.grad(u)
    )

def sigma(u):
    return lambda_ * ufl.nabla_div(u) * ufl.Identity(len(u)) + 2*mu*epsilon(u)

u = ufl.TrialFunction(V)
v = ufl.TestFunction(V)
f = fem.Constant(domain, default_scalar_type((0, 0, -rho * g)))
a = ufl.inner(sigma(u), epsilon(v)) * ufl.dx
L = ufl.dot(f, v) * ufl.dx + ufl.dot(T, v) * ds

# Solve
problem = LinearProblem(
    a,
    L,
    bcs=[bc],
    petsc_options={"ksp_type":"preonly","pc_type":"lu"},
    petsc_options_prefix="linear_elasticity",
)
uh = problem.solve()


# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

# Attach vector values to grid and warp grid by vector
grid["u"] = uh.x.array.reshape((geometry.shape[0], 3))
actor_0 = p.add_mesh(grid, style="wireframe", color="k")
warped = grid.warp_by_vector("u", factor=1)
actor_1 = p.add_mesh(warped, show_edges=True)
p.show_axes()
if not pyvista.OFF_SCREEN:
    p.show()
else:
    figure_as_array = p.screenshot("deflection.png") 

```

```python
with io.XDMFFile(domain.comm, "deformation.xdmf", "w") as xdmf:
    xdmf.write_mesh(domain)
    uh.name = "Deformation"
    xdmf.write_function(uh)
```

```python
# Stress computations
s = sigma(uh) - 1.0 / 3 * ufl.tr(sigma(uh)) * ufl.Identity(len(uh))
von_Mises = ufl.sqrt(3.0 / 2 * ufl.inner(s, s))

V_von_mises = fem.functionspace(domain, ("DG", 0))
stress_expr = fem.Expression(von_Mises, V_von_mises.element.interpolation_points)
stresses = fem.Function(V_von_mises)
stresses.interpolate(stress_expr)

warped.cell_data["VonMises"] = stresses.x.petsc_vec.array
warped.set_active_scalars("VonMises")
p = pyvista.Plotter()
p.add_mesh(warped)
p.show_axes()
if not pyvista.OFF_SCREEN:
    p.show()
else:
    stress_figure = p.screenshot("stresses.png")
```

### Bleyer Tutorial

```python
import pyvista
import numpy as np
from ufl import sym, grad, Identity, tr, inner, Measure, TestFunction, TrialFunction

from mpi4py import MPI

from dolfinx import fem, io, plot
import dolfinx.fem.petsc
from dolfinx.mesh import create_rectangle, CellType

length, height = 10, 1.0
Nx, Ny = 50, 5
domain = create_rectangle(
    MPI.COMM_WORLD,
    [np.array([0,0]), np.array([length, height])],
    [Nx, Ny],
    cell_type=CellType.quadrilateral,
)
dim = domain.topology.dim
print(f"Mesh topology dimension d={dim}")

degree = 2
shape = (dim,)
V = fem.functionspace(domain, ("P", degree, shape))

u_sol = fem.Function(V, name="Displacement")

E = fem.Constant(domain, 210e3)
nu = fem.Constant(domain, 0.3)

lmbda = E * nu / (1 + nu) / (1 - 2 * nu)
mu = E / 2 / (1 + nu)

def epsilon(v):
    return sym(grad(v))

def sigma(v):
    return lmbda * tr(epsilon(v)) * Identity(dim) + 2 * mu * epsilon(v)

print("mu (UFL):\n", mu)
print("epsilon (UFL): \n", epsilon(u_sol))
print("sigma (UFL):\n", sigma(u_sol))

u = TrialFunction(V)
v = TestFunction(V)

rho = 2e-3
g = 9.81
f = fem.Constant(domain, np.array([0, -rho * g]))

dx = Measure("dx", domain=domain)
a = inner(sigma(u), epsilon(v)) * dx
L = inner(f, v) * dx

def left(x):
    return np.isclose(x[0], 0)

def right(x):
    return np.isclose(x[0], length)

left_dofs = fem.locate_dofs_geometrical(V, left)
right_dofs = fem.locate_dofs_geometrical(V, right)

bcs = [
    fem.dirichletbc(np.zeros((2,)), left_dofs, V),
    fem.dirichletbc(np.zeros((2,)), right_dofs, V)
]

problem = fem.petsc.LinearProblem(
    a,
    L,
    u=u_sol,
    bcs=bcs,
    petsc_options={"ksp_type":"preonly", "pc_type":"lu"},
    petsc_options_prefix="linear_elasticity",
)
problem.solve()

vtk = io.VTKFile(domain.comm, "linear_elasticity.pvd", "w")
vtk.write_function(u_sol)
vtk.close()

# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

grid["u"] = np.c_[u_sol.x.array.reshape((-1,2)), np.zeros(geometry.shape[0])]
warped = grid.warp_by_vector("u", factor=500)
p.add_mesh(warped, show_edges=True)
p.view_xy()          
p.enable_parallel_projection()  
p.show_axes()
p.show()
```

### Adding a vertical force

```python
import pyvista

import numpy as np
from ufl import sym, grad, Identity, tr, inner, Measure, TestFunction, TrialFunction

import ufl
from dolfinx import mesh, plot
from dolfinx.mesh import meshtags

from mpi4py import MPI

from dolfinx import fem, io
import dolfinx.fem.petsc
from dolfinx.mesh import create_rectangle, CellType

length, height = 10, 1.0
Nx, Ny = 50, 5
domain = create_rectangle(
    MPI.COMM_WORLD,
    [np.array([0,0]), np.array([length, height])],
    [Nx, Ny],
    cell_type=CellType.quadrilateral,
)
dim = domain.topology.dim
print(f"Mesh topology dimension d={dim}")

degree = 2
shape = (dim,)
V = fem.functionspace(domain, ("P", degree, shape))

u_sol = fem.Function(V, name="Displacement")

E = fem.Constant(domain, 210e3)
nu = fem.Constant(domain, 0.3)

lmbda = E * nu / (1 + nu) / (1 - 2 * nu)
mu = E / 2 / (1 + nu)

def epsilon(v):
    return sym(grad(v))

def sigma(v):
    return lmbda * tr(epsilon(v)) * Identity(dim) + 2 * mu * epsilon(v)

u = TrialFunction(V)
v = TestFunction(V)

rho = 2e-3
g = 9.81
f = fem.Constant(domain, np.array([0, -rho * g]))

def left(x):
    return np.isclose(x[0], 0)

def top(x):
    return np.isclose(x[1], height)

# Pressure
fdim = domain.topology.dim - 1
top_facets = mesh.locate_entities_boundary(domain, fdim, top)

mt = meshtags(
    domain,
    fdim,
    top_facets,
    np.full(len(top_facets), 1, dtype=np.int32),
)

p = fem.Constant(domain, 10e-3)  
n = ufl.FacetNormal(domain)
T =  -p * n

# Linear problem
dx = Measure("dx", domain=domain)
ds = Measure("ds", domain=domain, subdomain_data=mt)
a = inner(sigma(u), epsilon(v)) * dx
L = inner(f, v) * dx + inner(T, v) * ds(1)

left_dofs = fem.locate_dofs_geometrical(V, left)

bcs = [
    fem.dirichletbc(np.zeros((2,)), left_dofs, V),
]

problem = fem.petsc.LinearProblem(
    a,
    L,
    u=u_sol,
    bcs=bcs,
    petsc_options={"ksp_type":"preonly", "pc_type":"lu"},
    petsc_options_prefix="linear_elasticity",
)
problem.solve()


vtk = io.VTKFile(domain.comm, "linear_elasticity.pvd", "w")
vtk.write_function(u_sol)
vtk.close()

# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

grid["u"] = np.c_[u_sol.x.array.reshape((-1,2)), np.zeros(geometry.shape[0])]
warped = grid.warp_by_vector("u", factor=500)
p.add_mesh(warped, show_edges=True)
p.view_xy()          
p.enable_parallel_projection()  
p.show_axes()
p.show()
```

### Tunnel geometry

```python
import pyvista
from mpi4py import MPI
from dolfinx.io import gmsh as gmshio
import gmsh
from dolfinx import plot

L, R = 50, 5
h = 1

gdim = 2
fdim = 1

gmsh.initialize()

occ = gmsh.model.occ
mesh_comm = MPI.COMM_WORLD
model_rank = 0
if mesh_comm.rank ==model_rank:
    plate = occ.addRectangle(0.0, 0.0, 0.0, L, L) # (x, y, z, dx, dy)
    notch = occ.addDisk(0.0, 0.0, 0.0, R, R) # (x, y, z, Rx, Ry=Rx)

    # cut 'plate' with notch
    outDimTags, _ = occ.cut([(gdim, plate)], [(gdim, notch)])
    # get tag of newly created object
    perf_plate = outDimTags[0][1]

    occ.synchronize()

    # identify boundary tags from bounding box
    eps = 1e-3 # tolerance for bounda=ing box
    left = gmsh.model.getEntitiesInBoundingBox(
        -eps, -eps, -eps, eps, L+eps, eps, dim=fdim
    )[0][1]
    bottom = gmsh.model.getEntitiesInBoundingBox(
        -eps, -eps, -eps, L+eps, eps, eps, dim=fdim
    )[0][1]
    right = gmsh.model.getEntitiesInBoundingBox(
        L-eps, -eps, -eps, L+eps, L+eps, eps, dim=fdim
    )[0][1]
    top = gmsh.model.getEntitiesInBoundingBox(
        -eps, L-eps, -eps, L+eps, L+eps, eps, dim=fdim
    )[0][1]

    # tag physical domains and facets
    gmsh.model.addPhysicalGroup(gdim, [perf_plate], 1)
    gmsh.model.addPhysicalGroup(fdim, [bottom], 1)
    gmsh.model.addPhysicalGroup(fdim, [right], 2)
    gmsh.model.addPhysicalGroup(fdim, [top], 3)
    gmsh.model.addPhysicalGroup(fdim, [left], 4)

    gmsh.option.setNumber("Mesh.CharacteristicLengthMin", h)
    gmsh.option.setNumber("Mesh.CharacteristicLengthMax", h)

    gmsh.model.mesh.generate(gdim)

    mesh_data = gmshio.model_to_mesh(gmsh.model, mesh_comm, model_rank, gdim=gdim)
    domain = mesh_data.mesh
    cell_tags = mesh_data.cell_tags
    facet_tags = mesh_data.facet_tags

    gmsh.finalize()
    
dim = domain.topology.dim
print(f"Mesh topology dimension d={dim}")  

pyvista.set_jupyter_backend("static")

topology, cell_types, geometry = plot.vtk_mesh(domain, gdim)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

# Create plotter and pyvista grid
p = pyvista.Plotter()
p.add_mesh(grid, show_edges=True)
p.view_xy()
p.show_axes()
p.show()
    
```

```python
from dolfinx import fem, io
import dolfinx.fem.petsc
import numpy as np
import ufl
from ufl import sym, grad, Identity, tr, inner, Measure, TestFunction, TrialFunction

# Material elastic parameter
E = fem.Constant(domain, 15000e6)
nu = fem.Constant(domain, 0.3)

lmbda = E * nu / (1 + nu) / (1 - 2 * nu)
mu = E / 2 / (1 + nu)

# Eigenstrains
V_strain = fem.functionspace(domain, ("DG", 0, (gdim, gdim)))
epsilon_star = fem.Function(V_strain, name="epsilon_star")
num_cells_local = domain.topology.index_map(gdim).size_local
strain_tensor = np.zeros((num_cells_local, gdim, gdim))

K = 0.75
sigma_v = 10e6
sigma_h = K * sigma_v
factor = (1 + nu.value) / E.value
eps_xx = factor * ((1 - nu.value) * sigma_h - nu.value * sigma_v)
eps_yy = factor * ((1 - nu.value) * sigma_v - nu.value * sigma_h)

strain_tensor[:, 0, 0] = eps_xx # in x direction
strain_tensor[:, 1, 1] = eps_yy
epsilon_star.x.petsc_vec[:] = strain_tensor.reshape(-1,)

# Define variational problem
def epsilon(v):
    return sym(grad(v))

def sigma(v, epsilon_star):
    epsilon_tot = epsilon(v) - epsilon_star
    return lmbda * tr(epsilon_tot) * Identity(gdim) + 2 * mu * (epsilon_tot)

# Web version uses lhs and rhs 
V = fem.functionspace(domain, ("P", 2, (gdim,)))
u = TrialFunction(V) 
v = TestFunction(V) 
Wint = ufl.inner(sigma(u, epsilon_star), epsilon(v)) * ufl.dx
aM = ufl.lhs(Wint)

# Uniform traction on top and right boundaries
n = ufl.FacetNormal(domain)
ds = Measure("ds", domain=domain, subdomain_data=facet_tags)

# Top traction
T_top = fem.Constant(domain, -sigma_v)
LM_top = ufl.dot(T_top*n, v) * ds(3)

# Right traction
T_right = fem.Constant(domain, -sigma_h)
LM_right = ufl.dot(T_right*n, v) * ds(2)

# Total L
LM = ufl.rhs(Wint) + LM_top + LM_right

# Dirichlet displacement bottom and left
V_ux, _ = V.sub(0).collapse()
V_uy, _ = V.sub(1).collapse()
bottom_dofs = fem.locate_dofs_topological((V.sub(1), V_uy), fdim, facet_tags.find(1))
left_dofs = fem.locate_dofs_topological((V.sub(0), V_ux), fdim, facet_tags.find(4))

ux0 = fem.Function(V_ux)
uy0 = fem.Function(V_uy)
bcs = [
    fem.dirichletbc(ux0, left_dofs, V.sub(0)),
    fem.dirichletbc(uy0, bottom_dofs, V.sub(1))
]

u_sol = fem.Function(V, name="Displacement")
problem = fem.petsc.LinearProblem(
    aM, 
    LM, 
    u=u_sol, 
    bcs=bcs, 
    petsc_options={"ksp_type":"preonly", "pc_type":"lu"},
    petsc_options_prefix="linear_elasticity_tunnel"
)
problem.solve()

######
vtk = io.VTKFile(domain.comm, "linear_elasticity_tunnel.pvd", "w")
vtk.write_function(u_sol)
vtk.close()

# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

grid["u"] = np.c_[u_sol.x.array.reshape((-1,2)), np.zeros(geometry.shape[0])]
warped = grid.warp_by_vector("u", factor=0)
p.add_mesh(warped, show_edges=True)
p.view_xy()          
p.enable_parallel_projection()  
p.show_axes()
p.show()
```

```python
import pyvista
from mpi4py import MPI
from dolfinx.io import gmsh as gmshio
import gmsh
from dolfinx import plot

L, H, R, d = 40, 25, 2, 5
h = 1

gdim = 2
fdim = 1

gmsh.initialize()

occ = gmsh.model.occ
mesh_comm = MPI.COMM_WORLD
model_rank = 0
if mesh_comm.rank==model_rank:

    # Box
    plate = occ.addRectangle(-L/2, -H, 0.0, L, H) # (x, y, z, dx, dy)
    
    # Cavity wall
    notch = occ.addDisk(0.0, -d, 0.0, R, R) # (x, y, z, Rx, Ry=Rx)

    # cut 'plate' with notch
    outDimTags, _ = occ.cut([(gdim, plate)], [(gdim, notch)])
    
    # # get tag of newly created object
    perf_plate = outDimTags[0][1]

    occ.synchronize()

    # identify boundary tags from bounding box
    eps = 1e-3 # tolerance for bounda=ing box
    left = gmsh.model.getEntitiesInBoundingBox(
        -L/2 -eps, -H -eps, -eps, -L/2+eps, eps, eps, dim=fdim
    )[0][1]
    bottom = gmsh.model.getEntitiesInBoundingBox(
        -L/2-eps, -H-eps, -eps, L/2+eps, -H+eps, eps, dim=fdim
    )[0][1]
    right = gmsh.model.getEntitiesInBoundingBox(
        L/2-eps, -H-eps, -eps, L/2+eps, eps, eps, dim=fdim
    )[0][1]
    top = gmsh.model.getEntitiesInBoundingBox(
        -L/2-eps, -eps, -eps, L/2+eps, eps, eps, dim=fdim
    )[0][1]
    
    # Régénérer avec le champ de taille
    gmsh.model.mesh.generate(gdim)

    # tag physical domains and facets
    gmsh.model.addPhysicalGroup(gdim, [perf_plate], 1)
    # gmsh.model.addPhysicalGroup(gdim, [plate], 1)
    gmsh.model.addPhysicalGroup(fdim, [bottom], 1)
    gmsh.model.addPhysicalGroup(fdim, [right], 2)
    gmsh.model.addPhysicalGroup(fdim, [top], 3)
    gmsh.model.addPhysicalGroup(fdim, [left], 4)

    gmsh.option.setNumber("Mesh.CharacteristicLengthMin", h)
    gmsh.option.setNumber("Mesh.CharacteristicLengthMax", h)

    gmsh.model.mesh.generate(gdim)

    mesh_data = gmshio.model_to_mesh(gmsh.model, mesh_comm, model_rank, gdim=gdim)
    domain = mesh_data.mesh
    cell_tags = mesh_data.cell_tags
    facet_tags = mesh_data.facet_tags

    gmsh.finalize()
    
dim = domain.topology.dim
print(f"Mesh topology dimension d={dim}")  

pyvista.set_jupyter_backend("static")

topology, cell_types, geometry = plot.vtk_mesh(domain, gdim)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

# Create plotter and pyvista grid
p = pyvista.Plotter()
p.add_mesh(grid, show_edges=True)
p.view_xy()
p.show_axes()
p.show()
    
```

```python
from dolfinx import fem, io, default_scalar_type
import dolfinx.fem.petsc
import numpy as np
import ufl
from ufl import sym, grad, Identity, tr, inner, Measure, TestFunction, TrialFunction

# Material elastic parameter
E = fem.Constant(domain, 15000e6)
nu = fem.Constant(domain, 0.3)

lmbda = E * nu / (1 + nu) / (1 - 2 * nu)
mu = E / 2 / (1 + nu)

# Gravity
gamma = 25e3 # N/m3

# Eigenstrains
V_strain = fem.functionspace(domain, ("DG", 0, (gdim, gdim)))
epsilon_star = fem.Function(V_strain, name="epsilon_star")
num_cells_local = domain.topology.index_map(gdim).size_local
strain_tensor = np.zeros((num_cells_local, gdim, gdim))

# # coordonnées des centres de cellules
num_cells_local = domain.topology.index_map(gdim).size_local
coords = dolfinx.mesh.compute_midpoints(domain, gdim, np.arange(num_cells_local))
y = coords[:,1]

sigma_v = - gamma * y
sigma_h = sigma_v

factor = (1 + nu.value) / E.value
eps_xx = factor * ((1 - nu.value) * sigma_h - nu.value * sigma_v)
eps_yy = factor * ((1 - nu.value) * sigma_v - nu.value * sigma_h)

strain_tensor[:,0,0] = eps_xx
strain_tensor[:,1,1] = eps_yy

epsilon_star.x.petsc_vec[:] = strain_tensor.reshape(-1,)

# Define variational problem
def epsilon(v):
    return sym(grad(v))

def sigma(v, epsilon_star):
    epsilon_tot = epsilon(v)  - epsilon_star
    return lmbda * tr(epsilon_tot) * Identity(gdim) + 2 * mu * (epsilon_tot)

# Web version uses lhs and rhs 
V = fem.functionspace(domain, ("P", 2, (gdim,)))
u = TrialFunction(V) 
v = TestFunction(V) 
Wint = ufl.inner(sigma(u, epsilon_star), epsilon(v)) * ufl.dx
aM = ufl.lhs(Wint)

# Traction free on top
T_top = fem.Constant(domain, default_scalar_type((0,0)))

# Volumic force
f = fem.Constant(domain, default_scalar_type((0, -gamma)))

# Non-uniform traction on left and right boundaries
ds = Measure("ds", domain=domain, subdomain_data=facet_tags)


LM = (ufl.rhs(Wint) + 
      ufl.dot(f, v) * ufl.dx + 
      ufl.dot(T_top, v) * ds(3) 
     )

# Dirichlet displacement bottom (ux and uy)
bottom_facets = facet_tags.find(1)
right_facets = facet_tags.find(2)
left_facets = facet_tags.find(4)

V_ux, _ = V.sub(0).collapse()
V_uy, _ = V.sub(1).collapse()
bottom_dofs = fem.locate_dofs_topological((V.sub(1), V_uy), fdim, bottom_facets)
left_dofs = fem.locate_dofs_topological((V.sub(0), V_ux), fdim, left_facets)
right_dofs = fem.locate_dofs_topological((V.sub(0), V_ux), fdim, right_facets)

ux0 = fem.Function(V_ux)
uy0 = fem.Function(V_uy)
bcs = [
    fem.dirichletbc(ux0, left_dofs, V.sub(0)),
    fem.dirichletbc(ux0, right_dofs, V.sub(0)),
    fem.dirichletbc(uy0, bottom_dofs, V.sub(1))
]

u_sol = fem.Function(V, name="Displacement")
problem = fem.petsc.LinearProblem(
    aM, 
    LM, 
    u=u_sol, 
    bcs=bcs, 
    petsc_options={"ksp_type":"preonly", "pc_type":"lu"},
    petsc_options_prefix="linear_elasticity_tunnel"
)
problem.solve()

##############
vtk = io.VTKFile(domain.comm, "linear_elasticity_utilidor.pvd", "w")
vtk.write_function(u_sol)
vtk.close()

# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

grid["u"] = np.c_[u_sol.x.array.reshape((-1,2)), np.zeros(geometry.shape[0])]
warped = grid.warp_by_vector("u", factor=0)
p.add_mesh(warped, show_edges=True)
p.view_xy()          
p.enable_parallel_projection()  
p.show_axes()
p.show()
```

```python
# Defining with lateral pressure for potentially having K not = to 1 but numerically unstable...
from dolfinx import fem, io, default_scalar_type
import dolfinx.fem.petsc
import numpy as np
import ufl
from ufl import sym, grad, Identity, tr, inner, Measure, TestFunction, TrialFunction

# Material elastic parameter
E = fem.Constant(domain, 15000e6)
nu = fem.Constant(domain, 0.3)

lmbda = E * nu / (1 + nu) / (1 - 2 * nu)
mu = E / 2 / (1 + nu)

# Gravity
gamma = 25e3 # N/m3

# Eigenstrains
V_strain = fem.functionspace(domain, ("DG", 0, (gdim, gdim)))
epsilon_star = fem.Function(V_strain, name="epsilon_star")
num_cells_local = domain.topology.index_map(gdim).size_local
strain_tensor = np.zeros((num_cells_local, gdim, gdim))

# # coordonnées des centres de cellules
num_cells_local = domain.topology.index_map(gdim).size_local
coords = dolfinx.mesh.compute_midpoints(domain, gdim, np.arange(num_cells_local))
y = coords[:,1]

K = 1
sigma_v = - gamma * y
sigma_h = K * sigma_v

factor = (1 + nu.value) / E.value
eps_xx = factor * ((1 - nu.value) * sigma_h - nu.value * sigma_v)
eps_yy = factor * ((1 - nu.value) * sigma_v - nu.value * sigma_h)

strain_tensor[:,0,0] = eps_xx
strain_tensor[:,1,1] = eps_yy

epsilon_star.x.petsc_vec[:] = strain_tensor.reshape(-1,)

# Define variational problem
def epsilon(v):
    return sym(grad(v))

def sigma(v, epsilon_star):
    epsilon_tot = epsilon(v)  - epsilon_star
    return lmbda * tr(epsilon_tot) * Identity(gdim) + 2 * mu * (epsilon_tot)

# Web version uses lhs and rhs 
V = fem.functionspace(domain, ("P", 2, (gdim,)))
u = TrialFunction(V) 
v = TestFunction(V) 
Wint = ufl.inner(sigma(u, epsilon_star), epsilon(v)) * ufl.dx
aM = ufl.lhs(Wint)

# Traction free on top
T_top = fem.Constant(domain, default_scalar_type((0,0)))

# Left and right tractions
V_traction = fem.functionspace(domain, ("P", 2))
sigma_h_func = fem.Function(V_traction)

def sigma_h_expr(x):
    return gamma * K * x[1]

sigma_h_func.interpolate(sigma_h_expr)

# Volumic force
f = fem.Constant(domain, default_scalar_type((0, -gamma)))

# Non-uniform traction on left and right boundaries
n = ufl.FacetNormal(domain)
ds = Measure("ds", domain=domain, subdomain_data=facet_tags)

LM = (ufl.rhs(Wint) + 
      ufl.dot(f, v) * ufl.dx + 
      ufl.dot(T_top, v) * ds(3) + 
      sigma_h_func * ufl.dot(n, v) * ds((2,4))
     )

# Dirichlet displacement bottom (ux and uy)
bottom_facets = facet_tags.find(1)

V_ux, _ = V.sub(0).collapse()
V_uy, _ = V.sub(1).collapse()
bottom_dofs = fem.locate_dofs_topological((V.sub(1), V_uy), fdim, bottom_facets)

ux0 = fem.Function(V_ux)
uy0 = fem.Function(V_uy)
bcs = [
    fem.dirichletbc(uy0, bottom_dofs, V.sub(1))
]

u_sol = fem.Function(V, name="Displacement")
problem = fem.petsc.LinearProblem(
    aM, 
    LM, 
    u=u_sol, 
    bcs=bcs, 
    petsc_options={"ksp_type":"preonly", "pc_type":"lu"},
    petsc_options_prefix="linear_elasticity_tunnel"
)
problem.solve()

##############
vtk = io.VTKFile(domain.comm, "linear_elasticity_utilidor.pvd", "w")
vtk.write_function(u_sol)
vtk.close()

# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

grid["u"] = np.c_[u_sol.x.array.reshape((-1,2)), np.zeros(geometry.shape[0])]
warped = grid.warp_by_vector("u", factor=0)
p.add_mesh(warped, show_edges=True)
p.view_xy()          
p.enable_parallel_projection()  
p.show_axes()
p.show()
```

### Utilidor mesh refinement

The problem under consideration is a circular shallow tunnel excavated in an elastic ground. To model the problem, it is assumed that the  vertical displacements are fixed at the bottom of the domain and that the horizontal displacements are fixed on the left and right sides of the domain. In the following, we define the domain as recangular, having a length $L = 40$ m and height $H = 30$ m. The tunnel has a radius $R = 2$ m and is located at depth $d = 4$ m from the surface.  

It is assumed that an in-situ state of stress rules in the massif. The vertical stress is due to gravity and the horizontal stress is assumed to be equal to the vertical stress. To model this in-situ stress state, eigenstrains can be incorporated in the formulation. Thus, the computed displacement corresponds only to the effects of excavation (disregarding the initial compression of the ground). 

The Hooke's law reads as:

$$\boldsymbol\sigma(\boldsymbol{u}) = \mathbb{C}: \left(\boldsymbol{\varepsilon}(\boldsymbol{u}) - \boldsymbol{\varepsilon}^{*})\right)$$

Using Lame's coefficients, one gets:

$$ \boldsymbol{\sigma}(\boldsymbol{u}) = \lambda \mathrm{tr} \left(\boldsymbol{\varepsilon}(\boldsymbol{u}) - \boldsymbol{\varepsilon}^{*}) \right) \boldsymbol{I} + 2\mu \left( \boldsymbol{\varepsilon}(\boldsymbol{u}) - \boldsymbol{\varepsilon}^{*} \right) $$

where $\boldsymbol{\sigma}(\boldsymbol{u})$ corresponds to the Cauchy stress tensor, $\boldsymbol{\varepsilon}(\boldsymbol{u})$ corresponds to the total deformations, where, assuming samll strains, are expressed as:

$$ \boldsymbol{\varepsilon}(\boldsymbol{u}) = \dfrac{1}{2} \left[\boldsymbol{\nabla}\boldsymbol{u} + (\boldsymbol{\nabla}\boldsymbol{u})^T)   \right] $$

The Hooke's law can then be written as:

\begin{equation}
\boldsymbol{\sigma}(\boldsymbol{u}) = \lambda \left[\nabla \cdot u - \mathrm{tr}(\boldsymbol{\varepsilon}^*) \right]\boldsymbol{I} + \mu \left[ \nabla \boldsymbol{u} + \left(\boldsymbol{u}\right)\right]
\end{equation}

$\boldsymbol{\varepsilon}^{*}$ denotes eigenstrains obtained in the presence of an in-situ stress state. Noting $\mathbb{S}$ the compliance tensor and $\boldsymbol{\sigma}^0$ th initial stress state ruling in the massif, eigenstrains are obtained following:

$$\boldsymbol{\varepsilon}^{*} = \mathbb{S}:\boldsymbol{\sigma}^0$$
where:
$$
\boldsymbol{\sigma}^0 =
\begin{bmatrix}
\sigma_h & 0 \\
0 & \sigma_v
\end{bmatrix}
$$
with $\sigma_v =\sigma_h = \gamma y$. 


In the presence of body forces (here gravity), the equilibrium equations is:

$$-\nabla\boldsymbol{\sigma}(\boldsymbol{u}) = \boldsymbol{f}$$

In FEniCS, one has to use the variational formulation to solve the problem. We refer here to for details. In this problem, the variational form reads as:

Find $\boldsymbol{u} \in V$, such that:


\begin{equation}
\int_\Omega \boldsymbol{\sigma}(\boldsymbol{u}) : \boldsymbol{\nabla}\boldsymbol{v} \, d\Omega
=
\int_\Omega \mathbf{f} \cdot \mathbf{v} \, d\Omega
+
\int_{\partial \Omega_T} \mathbf{T} \cdot \mathbf{v} \, ds,
\quad
\forall \mathbf{v} \in \mathbf{\hat V}
\end{equation}
where $\boldsymbol{u}$ is the unknown displacement (trial function) in the space of admissible displacements $\mathbf{V}$ and $\boldsymbol{v}$ corresponds to the test functions living in the space of admissible perturbations $\mathbf{\hat V}$. $\mathbf{T}$ corresponds to the surface traction (for instance surface load due to traffic or the presence of buildings). 

The left side of the equation is known as the bilinear form and often noted $a(\boldsymbol{u}, \boldsymbol{v})$ and the right term as the linear form noted $L(\boldsymbol{v})$ so that the variational form reads as:

\begin{equation}
a(\boldsymbol{u}, \boldsymbol{v}) = L(\boldsymbol{v})
\quad
\forall \mathbf{v} \in \mathbf{\hat V}
\end{equation}





It is possible to define the mesh directly in python by importing and using the `gmsh` library. 

```python
import gmsh

# Initialize GMSH
gmsh.initialize()
gmsh.model.add("utilidor")

# Geometrical parameters
L = 40.0      
H = 30.0      
d = 4
R = 2.0    

# Mesh size
mesh_size_far = 2.0     # Far away from tunnel wall
mesh_size_near = 0.2    # Close to the tunnel wall

# Rectangular box
p1 = gmsh.model.geo.addPoint(-L/2, -H, 0, mesh_size_far)
p2 = gmsh.model.geo.addPoint(L/2, -H, 0, mesh_size_far)
p3 = gmsh.model.geo.addPoint(L/2, 0, 0, mesh_size_far)
p4 = gmsh.model.geo.addPoint(-L/2, 0, 0, mesh_size_far)

l1 = gmsh.model.geo.addLine(p1, p2)
l2 = gmsh.model.geo.addLine(p2, p3)
l3 = gmsh.model.geo.addLine(p3, p4)
l4 = gmsh.model.geo.addLine(p4, p1)

outer_box = gmsh.model.geo.addCurveLoop([l1, l2, l3, l4])

# Tunnel cavity
center = gmsh.model.geo.addPoint(0, -d, 0, mesh_size_near)
p_tunnel_1 = gmsh.model.geo.addPoint(R, -d, 0, mesh_size_near)
p_tunnel_2 = gmsh.model.geo.addPoint(0, R-d, 0, mesh_size_near)
p_tunnel_3 = gmsh.model.geo.addPoint( -R, -d, 0, mesh_size_near)
p_tunnel_4 = gmsh.model.geo.addPoint(0, -R-d, 0, mesh_size_near)

arc1 = gmsh.model.geo.addCircleArc(p_tunnel_1, center, p_tunnel_2)
arc2 = gmsh.model.geo.addCircleArc(p_tunnel_2, center, p_tunnel_3)
arc3 = gmsh.model.geo.addCircleArc(p_tunnel_3, center, p_tunnel_4)
arc4 = gmsh.model.geo.addCircleArc(p_tunnel_4, center, p_tunnel_1)

tunnel_wall = gmsh.model.geo.addCurveLoop([arc1, arc2, arc3, arc4])

# Surface = domain - tunnel
surface = gmsh.model.geo.addPlaneSurface([outer_box, tunnel_wall])

gmsh.model.geo.synchronize()

# Define physical groups for further boundary conditions
gdim = 2
fdim = 1
gmsh.model.addPhysicalGroup(fdim, [l1], 1, name="bottom")
gmsh.model.addPhysicalGroup(fdim, [l2], 2, name="right")
gmsh.model.addPhysicalGroup(fdim, [l3], 3, name="top")
gmsh.model.addPhysicalGroup(fdim, [l4], 4, name="left")
gmsh.model.addPhysicalGroup(fdim, [arc1, arc2, arc3, arc4], 5, name="tunnel_wall")
gmsh.model.addPhysicalGroup(gdim, [surface], 6, name="domain")

# 2D mesh
gmsh.model.mesh.generate(gdim)

# Save gmsh file to visualize in Gmsh for instance
gmsh.write("tunnel_mesh.msh")
```

Then, we convert the mesh just created in `Gmsh` into a `dolfinx` mesh, and extract the physical markers (subdomains and boundaries) so we can use them in FEM simulations. This is done by importing and using `MPI` from `mpi4py` and `gmsh` from `dolfinx.io`.

```python
from mpi4py import MPI
from dolfinx.io import gmsh as gmshio

# Import in DOLFINx
mesh_comm = MPI.COMM_WORLD
model_rank = 0
mesh_data = gmshio.model_to_mesh(gmsh.model, mesh_comm, model_rank, gdim=gdim)
domain = mesh_data.mesh
cell_tags = mesh_data.cell_tags
facet_tags = mesh_data.facet_tags

# Terminates Gmsh API session
gmsh.finalize()

print(f"Number of cells: {domain.topology.index_map(gdim).size_global}")
```

Finally, we would like to visualize the mesh. This is done by importing the `pyvista` library and using the `plot` function from `dolfinx`.

```python
import pyvista
from dolfinx import plot

# Visualization
pyvista.set_jupyter_backend("static")

topology, cell_types, geometry = plot.vtk_mesh(domain, gdim)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

# Create plotter and pyvista grid
p = pyvista.Plotter()
p.add_mesh(grid, show_edges=True)
p.view_xy()
p.show_axes()
p.show()
```

Now, we aim to solve the boundary-value problem using the finite element solver FEniCS. Here, we define the material constants and apply an in-situ stress state.

```python
from dolfinx import fem
import dolfinx.fem.petsc
import numpy as np

# Material elastic parameter
E = fem.Constant(domain, 15000e6)
nu = fem.Constant(domain, 0.3)

lmbda = E * nu / (1 + nu) / (1 - 2 * nu)
mu = E / 2 / (1 + nu)

## Eigenstrains section

# Create a tensor to store the eigenstrain for each cell in the mesh
# The shape is (num_cells_local, gdim, gdim): for each cell, a gdim x gdim tensor is assigned (since we are in 2D gdim = 2)
num_cells_local = domain.topology.index_map(gdim).size_local
eigenstrain_tensor = np.zeros((num_cells_local, gdim, gdim))

# Compute the coordinates of the cell centroids.
coords = dolfinx.mesh.compute_midpoints(domain, gdim, np.arange(num_cells_local))
y = coords[:, 1]  

# Define the in-situ stress using based on gravity
gamma = 25e3  # N/m^3
sigma_v = -gamma * y  
sigma_h = sigma_v     

# Compute eigenstrains under the plane-strain assumption 
factor = (1 + nu.value) / E.value
eps_xx_star = factor * ((1 - nu.value) * sigma_h - nu.value * sigma_v)  
eps_yy_star = factor * ((1 - nu.value) * sigma_v - nu.value * sigma_h)  

# Assign the computed eigenstrain components to the strain tensor for each cell
eigenstrain_tensor[:, 0, 0] = eps_xx_star  
eigenstrain_tensor[:, 1, 1] = eps_yy_star  

# Define a function space for the eigenstrain field
# "DG" (Discontinuous Galerkin) of order 0 is chosen because eigenstrains may be discontinuous across elements
V_strain = fem.functionspace(domain, ("DG", 0, (gdim, gdim)))

# Create a function in this space to represent the eigenstrain tensor field
epsilon_star = fem.Function(V_strain, name="epsilon_star")

# Assign values from a the precomputed eigenstrain tensor to the function
# The tensor is reshaped into a 1D array to match the PETSc vector layout of the function
epsilon_star.x.petsc_vec[:] = eigenstrain_tensor.reshape(-1,)
```

We now formulate the variational problem. This is done in FEniCS by importing the `UFL` (Unified Form Language) package that allows us to formalize the symbolic expressions of the constitutive behavior. 

In FEniCS, one can use the UFL functions `ufl.lhs` and `ufl.rhs` to extract the bilinear and respectivelely the linear terms (separating the terms that depend on both the trial and test functions on the one side and the terms that only depend on the test function on the other).


```python
import ufl
from ufl import sym, grad, Identity, tr, inner, Measure, TestFunction, TrialFunction
from dolfinx import fem, io, default_scalar_type

# Small strains
def epsilon(v):
    return sym(grad(v))

# Constitutive bahavior
def sigma(v, epsilon_star):
    epsilon_tot = epsilon(v) - epsilon_star
    return lmbda * tr(epsilon_tot) * Identity(gdim) + 2 * mu * (epsilon_tot)

# Creation of a function space for both the Trial and Test functions. 
# Note that the same space is used for both. 
# "P", 2 : quadratic Lagrange elements (degree 2)
# (gdim,) : 2 components for u_x and u_y displacements
V = fem.functionspace(domain, ("P", 2, (gdim,)))
u = TrialFunction(V) 
v = TestFunction(V) 
u_sol = fem.Function(V, name="Displacement")

# Left part of the variational formulation
# Define domain integration measure (by default ufl.dx would work)
dx = ufl.Measure("dx", domain=domain)
Wint = ufl.inner(sigma(u, epsilon_star), epsilon(v)) * dx

# Extract the bilinear term 
aM = ufl.lhs(Wint)

# A Measure is created to allow integrating over facets (boundary surfaces) 
ds = Measure("ds", domain=domain, subdomain_data=facet_tags)

# Traction-free on top - To be modified if necessary
T_top = fem.Constant(domain, default_scalar_type((0,0)))

# Volumetric body force 
f = fem.Constant(domain, default_scalar_type((0, -gamma)))

# Linear term (right part of the variational form)
LM = (ufl.rhs(Wint) + 
      ufl.dot(f, v) * ufl.dx + 
      ufl.dot(T_top, v) * ds(3) 
     )

# Dirichlet - Fixed vertical displacement at the bottom (uy) and horizontal left/right (ux)
bottom_facets = facet_tags.find(1)
right_facets = facet_tags.find(2)
left_facets = facet_tags.find(4)

# Extract function spaces for each displacement component separately
# .collapse() gives the underlying scalar function space needed for applying Dirichlet BCs.
V_ux, _ = V.sub(0).collapse()
V_uy, _ = V.sub(1).collapse()

# Locate the degrees of freedom (DoFs) associated with each boundary
bottom_dofs = fem.locate_dofs_topological((V.sub(1), V_uy), fdim, bottom_facets)
left_dofs = fem.locate_dofs_topological((V.sub(0), V_ux), fdim, left_facets)
right_dofs = fem.locate_dofs_topological((V.sub(0), V_ux), fdim, right_facets)

# Functions defining the Dirichlet values (zero displacement by default)
ux0 = fem.Function(V_ux)
uy0 = fem.Function(V_uy)

# Define the actual Dirichlet boundary conditions
bcs = [
    fem.dirichletbc(ux0, left_dofs, V.sub(0)),
    fem.dirichletbc(ux0, right_dofs, V.sub(0)),
    fem.dirichletbc(uy0, bottom_dofs, V.sub(1))
]

# Wrap-up the problem
problem = fem.petsc.LinearProblem(
    aM, 
    LM, 
    u=u_sol, 
    bcs=bcs, 
    petsc_options={"ksp_type":"preonly", "pc_type":"lu"},
    petsc_options_prefix="linear_elasticity_utilidor"
)
problem.solve()

# Create pvd file for using in Paraview
vtk = io.VTKFile(domain.comm, "linear_elasticity_utilidor.pvd", "w")
vtk.write_function(u_sol)
vtk.close()

# Visualisation
p = pyvista.Plotter()
topology, cell_types, geometry = plot.vtk_mesh(V)
grid = pyvista.UnstructuredGrid(topology, cell_types, geometry)

grid["u"] = np.c_[u_sol.x.array.reshape((-1,2)), np.zeros(geometry.shape[0])]
warped = grid.warp_by_vector("u", factor=0)
p.add_mesh(warped, show_edges=True)
p.view_xy()          
p.enable_parallel_projection()  
p.show_axes()
p.show()
```

We can now define functions to visualise the stress - Here Von Mises

```python
# Stress computations
def stress_VM(u):
    s = lmbda * tr(epsilon(u)) * Identity(gdim) + 2 * mu * (epsilon(u))
    s_dev = s - (tr(s) / gdim) * Identity(gdim)  # Déviateur
    von_Mises = ufl.sqrt(3.0 / 2 * ufl.inner(s_dev, s_dev))
    return von_Mises

von_Mises = stress_VM(u_sol)

V_von_mises = fem.functionspace(domain, ("DG", 0))
stress_expr = fem.Expression(von_Mises, V_von_mises.element.interpolation_points)
stresses = fem.Function(V_von_mises)
stresses.interpolate(stress_expr)

warped.cell_data["VonMises"] = stresses.x.petsc_vec.array
warped.set_active_scalars("VonMises")
p = pyvista.Plotter()
p.add_mesh(warped)
p.view_xy()   
p.show_axes()
if not pyvista.OFF_SCREEN:
    p.show()
else:
    stress_figure = p.screenshot("Von_Mises_Stress.png")
```
